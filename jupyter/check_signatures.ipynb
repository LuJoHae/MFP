```python
from pathlib import Path
import pandas as pd
from pprint import pprint
```

```python
signature_dir = Path("../signatures").resolve()
assert signature_dir.exists() and signature_dir.is_dir()
```

```python
pd.read_csv(signature_dir / "gene_signatures_order.tsv", sep="\t", header=None).rename(columns={0: "signature"})
```

```python
def read_gmt(file_path):
    with open(file_path, 'r') as f:
        return {line.split('\t')[0]: {"name": line.split('\t')[1], "genes": line.strip().split('\t')[2:]} for line in f}

pathways = read_gmt(signature_dir / "gene_signatures.gmt")
pprint(pathways)
```
